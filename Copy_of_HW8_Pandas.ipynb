**4. Описова статистика**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній розмір рахунку на всьому наборі даних

- Кількість пропущених значень у всіх колонках

```python
average_bill = df['total_bill'].mean()
print( average_bill) #середній розмір рахунку
max_tip = df['tip'].max()
print("Максимальні чайові:", max_tip)
missing_values = df.isnull().sum()
print("Кількість пропущених значень у кожній колонці:")
print(missing_values)
meal_counts = df['time'].value_counts()
print("Кількість рахунків за обіди і вечері:")
print(meal_counts)
```

**5. Статистика із групуванням**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній рахунок за статтю (groupby `sex`)
- Середні чайові за днем тижня (groupby `day`)
- Скільки відвідувачів у кожен день

```python
avg_bill_by_sex = df.groupby('sex')['total_bill'].mean()
print("Середній рахунок за статтю:")
print(avg_bill_by_sex)
print("\n")
avg_tip_by_day = df.groupby('day')['tip'].mean()
print("Середні чайові за днем тижня:")
print(avg_tip_by_day)
print("\n")
visitors_by_day = df['day'].value_counts()
print("Кількість відвідувачів у кожен день:")
print(visitors_by_day)
```
