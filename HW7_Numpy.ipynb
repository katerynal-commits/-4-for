**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
import numpy as np  # імпорт бібліотеки NumPy
arr = np.array([5, 15, 25, 35, 45])
print(arr) #масив зі списку
zeros_arr = np.zeros(7, dtype=int)
print(zeros_arr) #масив з нулів розміром 7
ones_arr = np.ones(6, dtype=int)
print(ones_arr) #масив одиниць розміром 6
range_arr = np.arange(10, 20)
print(range_arr)  #масив від 10 до 19
```

**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
import numpy as np  # імпорт бібліотеки NumPy
arr = np.array([12, 24, 36, 48, 60])
multiplied = arr * 2
print(multiplied) #Х2
added = arr + 10
print(added)  # +10
squared = arr ** 2
print(squared) # цифри у квадраті
sqrt_arr = np.sqrt(arr)
print(sqrt_arr) # квадратний корінь з усіх елементів
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
import numpy as np  # імпорт бібліотеки NumPy
arr = np.array([18, 27, 33, 41, 52, 68, 79])
print(arr)
mean_value = np.mean(arr)
print(mean_value)  #середнє значення
min_value = np.min(arr)
print(min_value) #мінімум
max_value = np.max(arr)
print(max_value) #максимум
sum_value = np.sum(arr)
print(sum_value)  #сума
```
